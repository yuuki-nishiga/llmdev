```python
grades = [70, 75, 80, 85, 90]
average = 0
above_average_score = 0

for i in range(len(grades)):
    average += grades[i]
print(f"平均点: {average/len(grades)}")
print(f"最高点: {max(grades)}")
print(f"最低点: {min(grades)}")

for j in range(len(grades)):
    if grades[j] >= average/len(grades):
        above_average_score += 1
print(f"平均以上の学生数: {above_average_score}")
```
