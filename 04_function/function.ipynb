```python
def is_palindrome(s):
    if s == s[::-1]:
        print(f"「{s}」は回文です！")
    else:
        print(f"「{s}」は回文ではないです。")

def greet(name):
    print(f"こんにちは、{name}さん！")

def introduce(name, age=18):
    print(f"こんにちは、{name}さん！ 年齢は{age}歳ですね。")

def add(a, b):
    result = a + b
    return result

def introduce_dict(**kwargs):
    for key, value in kwargs.items():
        print(f"{key} : {value}")

def calculate(a, b):
    sum_result = a + b
    product_result = a * b
    return sum_result, product_result

def greet_none():
    print("Hello!")

def multiply(a, b):
    return a * b

# 2乗を行う関数
def square(number):
    return multiply(number, number)

is_palindrome("にんてんどううどんてんに")

is_palindrome("Pythonは便利だが遅い")

is_palindrome("no devil lived on")

greet("太郎")
greet("花子")

introduce("太郎", 20)
introduce("花子")

total = add(3, 5)
print(total)

introduce_dict(name="太郎", age=20, hobby="読書")

result1, result2 = calculate(3, 4)
print("足し算結果:", result1)
print("掛け算結果:", result2)

result = greet_none()
print(result)

print(square(5))
```
