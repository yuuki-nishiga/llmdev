```python
def username_judge(name):
    if len(name) < 5 or len(name) > 15:
        print("エラー: ユーザー名は5文字以上15文字以内で入力してください")
        return False
    elif " " in name:
        print("エラー: ユーザー名に空白を含めることはできません")
        return False
    elif name.isalnum() == False:
        print("エラー: ユーザー名は英数字のみ使用できます")
        return False
    
    return True

username = input("ユーザー名を入力してください")
if username_judge(username) == True:
    print("ユーザー名は有効です")
else:
    print("不正なユーザー名です")
```
