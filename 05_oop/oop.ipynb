```python
class Animal:
    def speak(self):
        return "何か音を出しています。"

class Dog(Animal):
    def speak(self):
        return "ワンワン！"

class Cat(Animal):
    def speak(self):
        return "ニャー！"

animals = [Dog(), Cat()]

for animal in animals:
    print(animal.speak())  # 各オブジェクトに応じたメソッドが呼ばれる
```
