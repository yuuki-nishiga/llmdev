```python
class BankAccount:
    def __init__(self, balance:int):
        self.__balance = balance
    
    @property
    def get_balance(self):
        return self.__balance

    def deposit(self, amount):
        if amount <= 0:
            print("エラー: 正しい金額を入力してください")
        else:
            self.__balance += amount

    def withdraw(self, amount:int):
        if self.__balance < amount or amount <= 0:
            print("エラー: 残高不足または無効な金額です")
        else:
            self.__balance -= amount
    
balance = int(input("初期残高を設定してください: "))
print(f"初期残高: {balance}")
bankAccount = BankAccount(balance)
while True:
    print("番号を入力してください[1:預け入れ, 2:引き出し, 3:残高表示 4:終了]:", end="", flush=True)
    choice = input("")

    if choice == "1":
        amount = int(input("預け入れする金額を入力してください。"))
        print(f"預け入れ: {amount}")
        bankAccount.deposit(amount)
    
    elif choice == "2":
        amount = int(input("引き出しする金額を入力してください。"))
        print(f"引き出し: {amount}")
        bankAccount.withdraw(amount)

    elif choice == "3":
        print(bankAccount.get_balance)

    elif choice == "4":
        print("銀行口座管理を終了します。")
        break

    else:
        print("1-4の数字を入力してください。", flush=True)
```
