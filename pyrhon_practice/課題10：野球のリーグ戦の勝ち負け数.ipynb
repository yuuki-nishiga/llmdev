```python
class Baseball_Team:
    def __init__(self, name, win, lose, draw):
        self.name = name
        self.win = win
        self.lose = lose
        self.draw = draw
    
    def calc_win_rate(self):
        return self.win / (self.win + self.lose)
    
    def show_team_result(self):
        print(f"{self.name} {self.win} {self.lose} {self.draw} {self.calc_win_rate():.3f}")
        return ""
    
```

```python
baseball1 = Baseball_Team("Archers", 77, 64, 2)
baseball2 = Baseball_Team("Meteors", 71, 69, 3)
baseball3 = Baseball_Team("Cheetahs", 69, 68, 6)
baseball4 = Baseball_Team("Sharks", 70, 70, 3)
baseball5 = Baseball_Team("Pythons", 68, 73, 2)
baseball6 = Baseball_Team("Owls", 59, 82, 2)
print("team win rose draw rate")
print(baseball1.show_team_result())
print(baseball2.show_team_result())
print(baseball3.show_team_result())
print(baseball4.show_team_result())
print(baseball5.show_team_result())
print(baseball6.show_team_result())
```
