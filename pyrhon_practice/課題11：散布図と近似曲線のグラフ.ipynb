```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([62, 95, 83, 100, 72, 71, 55, 85, 67, 75])   # english
y = np.array([62, 80, 85, 95, 58, 76, 52, 84, 67, 71])    # math

p = np.polyfit(x, y, 1)

f = np.poly1d(p)

plt.scatter(x, y, label="score")
plt.plot(x, f(x), color="red", label="lsm")
plt.legend()
plt.xlabel("english")
plt.ylabel("math")

plt.show()
```
