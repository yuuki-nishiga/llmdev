```python
grades = []
average = 0

while True:
    score = int(input("点数を入力してください"))
    print(f"点数を入力してください：{score}")
    if score == -1:
        break
    else:
        grades.append(score)
        

for i in range(len(grades)):
    average += grades[i]
print(f"{len(grades)}人のテストの平均点は{average/len(grades)}です")
```
