```python
try:
    num1 = int(input("割られる数を入力してください"))
    print(f"割られる数を入力してください: {num1}")
    num2 = int(input("割る数を入力してください"))
    print(f"割る数を入力してください: {num2}")
    result = num1 / num2
except ZeroDivisionError:
    print("エラー: 0で割り算しないでください")
except ValueError:
    print("エラー: 数値を入力してください")
else:
    print(f"{num1} ÷ {num2} = {result}")
print("処理を終了します")
```
