```python
words_list = []
keys = ["a", "b", "c", "d", "e", "f", "g", 
        "h", "i", "j", "k", "l", "m", "n",
        "o", "p", "q", "r", "s", "t", "u",
        "v", "w", "x", "y", "z"]
alphabet_count = dict.fromkeys(keys, 0)

while True:
    input_word = input("英単語を入力してください")
    if input_word == "":
        break
    print(f"英単語を入力してください：{input_word}")
    words_list.append(input_word)
    for char in input_word:
        alphabet_count[char] += 1

    

words_list.sort()
print(f"入力した英単語: {words_list}")

for key, value in alphabet_count.items():
    if value > 0:
        print(f"{key}が{value}個ありました")
```
